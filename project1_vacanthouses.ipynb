```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www8.cao.go.jp/kourei/whitepaper/w-2022/html/zenbun/s1_1_4.html'
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html, 'html.parser')

table = soup.find('table')
```

```python
data = []
for row in table.find_all('tr'):
    cols = row.find_all('td')  
    cols = [ele.text.strip() for ele in cols]  
    data.append([ele for ele in cols if ele])  
new_data = data[2:-1]
new_data
```

```python
df_kourei = pd.DataFrame(new_data, columns=['総人口（2021年）', '65歳以上人口（2021年）', '高齢化率（2021年）', '高齢化率（2045年）', '高齢化率の伸び（ポイント）'])
df_kourei
```
